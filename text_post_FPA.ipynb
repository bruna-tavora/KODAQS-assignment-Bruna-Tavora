```python
# Instalar bibliotecas
!pip install bertopic sentence-transformers umap-learn hdbscan emoji -q

# Importar bibliotecas
import pandas as pd
import re
import emoji
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
from bertopic import BERTopic
from sentence_transformers import SentenceTransformer

# Carregar o dataset
df = pd.read_csv("texto_FPA.csv")
df.columns = df.columns.str.strip()
df = df.dropna(subset=['texto'])
```

```python


# Pr√©-processamento: limpeza de texto
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

# Combinar stopwords em portugu√™s e ingl√™s
stop_words = set(stopwords.words('portuguese') + stopwords.words('english'))

# Fun√ß√£o de limpeza com remo√ß√£o de stopwords multil√≠ngues
def clean_text(text):
    text = text.lower()
    text = emoji.replace_emoji(text, replace='')                      # remove emojis
    text = re.sub(r'[^a-z√°√©√≠√≥√∫√†√®√¨√≤√π√ß√£√µ√¢√™√Æ√¥√ª\s]', '', text)            # remove pontua√ß√£o/n√∫meros
    text = re.sub(r'\s+', ' ', text)                                  # remove espa√ßos duplos
    tokens = text.split()
    tokens = [word for word in tokens if word not in stop_words]      # remove stopwords multil√≠ngues
    return ' '.join(tokens)

# Aplicar a fun√ß√£o
df['cleaned_text'] = df['texto'].astype(str).apply(clean_text)
texts = df['cleaned_text'].tolist()

```

```python


# Modelo multil√≠ngue da Hugging Face
embedding_model = SentenceTransformer("paraphrase-multilingual-MiniLM-L12-v2")

# Inicializar e treinar BERTopic
topic_model = BERTopic(embedding_model=embedding_model, language="multilingual")
topics, probs = topic_model.fit_transform(texts)

# Visualiza√ß√£o
topic_model.visualize_barchart(top_n_topics=10)
```

```python


# Stopwords em portugu√™s + ingl√™s
base_stopwords = set(stopwords.words('portuguese') + stopwords.words('english'))


```

```python


# Stopwords personalizadas
custom_stopwords = {
    "pedro", "lupion", "ter√ßafeira", "nesta", "membros", "parlamentares", "agropecu√°ria",
    "parlamentar", "coletiva", "presidente", "agenciafpa", "sobre", "frente", "fpa", "neste"
    "nesta", "FPA" ""
}

# Adicionar palavras que correspondem a padr√µes com curingas
pattern_stopwords = ["deputad", "senador"]

# Fun√ß√£o de limpeza
def clean_text(text):
    text = text.lower()
    text = emoji.replace_emoji(text, replace='')
    text = re.sub(r'[^a-z√°√©√≠√≥√∫√†√®√¨√≤√π√ß√£√µ√¢√™√Æ√¥√ª\s]', '', text)
    text = re.sub(r'\s+', ' ', text)
    tokens = text.split()

    # Remover stopwords padr√£o + customizadas + prefixos com wildcard
    filtered = []
    for word in tokens:
        if word in base_stopwords or word in custom_stopwords:
            continue
        if any(word.startswith(prefix) for prefix in pattern_stopwords):
            continue
        filtered.append(word)

    return ' '.join(filtered)

# Aplicar ao dataframe
df['cleaned_text'] = df['texto'].astype(str).apply(clean_text)
texts = df['cleaned_text'].tolist()
```
